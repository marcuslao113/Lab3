```python
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, # <-- Note that Python does not complain about this extra comma. This is perfectly okay.
       #     The creators of Python saw that dictionaries tend to be changed a lot, so allowing this is a
       #     convenience for developers who need to add elements when needed.
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

import math
def getdistance(dest1, dest2):
    longitude1=locations[dest1]["longitude"]
    longitude2=locations[dest2]["longitude"]
    
    latitude1=locations[dest1]["latitude"]
    latitude2=locations[dest2]["latitude"]
    
    distance=math.sqrt(((longitude2-longitude1)**2) + ((latitude2-latitude1)**2))
    return distance

def eta(source,destination):
    
    # write your code here
    eta=0
    dest= []
    for i in locations:
        dest.append(i)
        
    loc1=source
    loc2=destination
    dest.remove(loc1)
    
    dest.remove(loc2)
    
    loc3= dest[0]
    
    #print(getdistance(loc1, loc2))
    #print(getdistance(loc1,loc3))
    if getdistance(loc1, loc2)>getdistance(loc1, loc3):
        
        
        eta=legs[loc1, loc2]["travel_time_mins"]
        
    else: 
        eta=legs[loc1, loc3]["travel_time_mins"]
        
        eta+=legs[loc3, loc2]["travel_time_mins"]
        
    return eta 

counter = 1
while counter < 3:
    if counter == 1:
        source=input("Enter source: ")
    
    elif counter ==2:
        destination=input("Enter destination: ")
    counter = counter + 1
    
print(eta(source, destination))


    
```
